```python
import pandas as pd
```

```python
def create_entries(file, entry_size):
    entries = []

    words = []
    f = open(file)
    for line in f:
        words.extend(line.split())
    
    entry = ""
    for i in range(len(words) - 1):
        if i != 0 and i % entry_size == 0:
            entries.append(entry.rstrip())
            entry = ""
        entry += words[i] + " "
    
    return len(entries), entries
```

```python
all_entries = []
all_labels = []

def create_series(file, label, entry_size):
    num_entries, entries = create_entries(file, entry_size)
    all_entries.extend(entries)
    all_labels.extend([label] * num_entries)

phils = ["absurdism", "epicureanism", "existentialism", "stoicism", "utilitarianism"]
for p in phils:
    create_series(f"data/{p}.txt", p, 300)

df = pd.DataFrame({
    "entry": all_entries,
    "label": all_labels
})
```

```python
df.to_csv("data/data.csv")
```
