## Nature of Things

```python
lines = []

with open("data/nature_of_things.txt") as f:
    for line in f:
        if line != "\n" and line != line.upper():
            lines.append(line.rstrip().lstrip())
```

```python
with open("data/clean_nature_of_things.txt", "w") as f:
    f.write("\n".join(lines))
```

```python
print(len(lines))
```
